```
%tensorflow_version 2.x
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import IPython.display as display
import numpy as np
print(f"Tensorflow version: {tf.__version__}")
tf.executing_eagerly()
```

```
ds_train, ds_valid, ds_test = tfds.load("cars196", split=["train[-80%:]", "train[:10%]", "test"])
for example in ds_train.take(5):
    bbox, image, label = example["bbox"], example["image"], example["label"]
    image = tf.image.resize(image, [300, 600]) / 255
    plt.imshow(image)
    plt.show()
    print(f"Bbox: {bbox}")
    print(f"Image size: {image.shape}, {image.dtype}")
    print(f"Label: {label}")
```

```
try:
  tpu = tf.distribute.cluster_resolver.TPUClusterResolver() # TPU detection
  print(f"Running on TPU {tpu.cluster_spec().as_dict()['worker']}")
except ValueError:
  raise BaseException("Error: not connected to TPU runtime")

tf.config.experimental_connect_to_cluster(tpu)
tf.tpu.experimental.initialize_tpu_system(tpu)
tpu_strategy = tf.distribute.experimental.TPUStrategy(tpu)
```
